```python
df_education = pd.read_csv('crime_education.csv', encoding='cp949')
df_education.head()
```

```python
df_job_cont = df_job[df_job.columns.difference(['범죄대분류', '범죄중분류'])]
```

```python
df_education_job = df_education.copy()
df_education_job[df_job_cont.columns] = df_job_cont.values
```

```python
df_education_job.info()
```

```python
df_education_job_trans = df_education_job.transpose(copy=True)
```

```python
df_education_job_trans.rename(columns=df_education_job_trans.iloc[1], inplace=True)
df_education_job_trans.drop(df_education_job_trans.index[0:2], inplace=True)
```

```python
df_education_job_trans[df_education_job_trans.columns] = df_education_job_trans[df_education_job_trans.columns].apply(pd.to_numeric)
```

```python
df_education_job_trans.head()
```

```python
df_education_job_trans.info()
```

```python
df_education_job_trans.drop(['기타', '미상'], axis=0, inplace=True)
```

```python
new_df_edu_job_trans = df_education_job_trans.copy()
```

```python
etc = df_education_job_trans.loc['기타(공익요원)':'무직자', : ].sum(axis=0)
etc.name = '기타(공익,전.의경,주부,학생)'
new_df_edu_job_trans = new_df_edu_job_trans.append(etc)
```

```python
new_df_edu_job_trans.drop(['기타(공익요원)', '기타(기타)', '기타(전경.의경)', '기타(주부)', '기타(학생)'], axis=0, inplace=True)
```

```python
smbusiness = df_education_job_trans.loc['자영업(건설업)':'전문직(교수)', : ].sum(axis=0)
smbusiness.name = '자영업'
new_df_edu_job_trans = new_df_edu_job_trans.append(smbusiness)
new_df_edu_job_trans.drop(['자영업(건설업)', '자영업(광업)', '자영업(금융업)', '자영업(기타사업)', '자영업(노점)',
       '자영업(농.임.수산업)', '자영업(도.소매업)', '자영업(무역업)', '자영업(부동산업)', '자영업(숙박업)',
       '자영업(요식업)', '자영업(유흥업)', '자영업(의료보건업)', '자영업(제조업)', '자영업(차량정비업)',
       '자영업(행상)'], axis=0, inplace=True)
```

```python
prof = df_education_job_trans.loc['전문직(교수)':'피고용자(경비원)', : ].sum(axis=0)
prof.name = '전문직'
new_df_edu_job_trans = new_df_edu_job_trans.append(prof)
new_df_edu_job_trans.drop(['전문직(교수)', '전문직(기타전문직)', '전문직(변호사)', '전문직(언론인)', '전문직(예술인)', '전문직(의사)',
       '전문직(종교가)'], axis=0, inplace=True)
```

```python
employee = df_education_job_trans.loc['피고용자(경비원)': , : ].sum(axis=0)
employee.name = '피고용자'
new_df_edu_job_trans = new_df_edu_job_trans.append(employee)
new_df_edu_job_trans.drop(['피고용자(경비원)', '피고용자(공원)', '피고용자(교원(사립))', '피고용자(국공영기업체직원)',
       '피고용자(금융기관직원)', '피고용자(기술자)', '피고용자(기타피고용자)', '피고용자(사무원)', '피고용자(외판원)',
       '피고용자(요식업종사자)', '피고용자(운전사)', '피고용자(유흥업종사자)', '피고용자(일반회사원)',
       '피고용자(일용노동자)', '피고용자(점원)'], axis=0, inplace=True)
```

```python
sc_drop = df_education_job_trans.loc[('초등학교(중퇴)', '중학교(중퇴)', '고등학교(중퇴)') , : ].sum(axis=0)
sc_drop.name = '초중고_중퇴자'
new_df_edu_job_trans = new_df_edu_job_trans.append(sc_drop)
new_df_edu_job_trans.drop(['초등학교(중퇴)', '중학교(중퇴)', '고등학교(중퇴)'], axis=0, inplace=True)
```

```python
teenagers = df_education_job_trans.loc[('초등학교(재중)', '초등학교(졸업)', '중학교(재중)', '중학교(졸업)', '고등학교(재중)') , : ].sum(axis=0)
teenagers.name = '10대_학생'
new_df_edu_job_trans = new_df_edu_job_trans.append(teenagers)
new_df_edu_job_trans.drop(['초등학교(재중)', '초등학교(졸업)', '중학교(재중)', '중학교(졸업)', '고등학교(재중)'], axis=0, inplace=True)
```

```python
univ_drop = df_education_job_trans.loc[('대학(4년미만)(중퇴)', '대학(4년이상)(중퇴)') , : ].sum(axis=0)
univ_drop.name = '대학중퇴자'
new_df_edu_job_trans = new_df_edu_job_trans.append(univ_drop)
new_df_edu_job_trans.drop(['대학(4년미만)(중퇴)', '대학(4년이상)(중퇴)'], axis=0, inplace=True)
```

```python
univ_stud = df_education_job_trans.loc[('대학(4년미만)(재중)', '대학(4년이상)(재중)', '대학원(재중)') , : ].sum(axis=0)
univ_stud.name = '대학(원)생'
new_df_edu_job_trans = new_df_edu_job_trans.append(univ_stud)
new_df_edu_job_trans.drop(['대학(4년미만)(재중)', '대학(4년이상)(재중)', '대학원(재중)'], axis=0, inplace=True)
```

```python
cols = new_df_edu_job_trans.columns[:-1]
rev_cols = list(reversed(cols))
fig, axs = plt.subplots(len(cols), 2, sharex=True, figsize=(15,200))

for i in range(len(cols)):
    for j in range(2):
        fig.suptitle("직업, 교육정도에 따른 범죄율")
        axs[i, j].set_title(cols[i]+"-"+rev_cols[i])
        x, y = pd.Series(new_df_edu_job_trans[cols[i]], name=cols[i]), pd.Series(new_df_edu_job_trans[rev_cols[i]], name=rev_cols[i])
        sns.regplot(x=x, y=y, marker="+", ax=axs[i, j])
```
